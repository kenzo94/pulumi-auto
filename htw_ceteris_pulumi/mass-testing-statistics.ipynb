```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
#rounded values
time = [18, 40, 30, 40, 38, 30, 35, 37, 18, 30]
iteration = np.arange(1,11)
iteration_label = ["9:00", "10:00", "12:00","14:00","15:00", "17:00","19:00", "20:00", "21:00", "22:00"]
```

```python
std = np.std(time) # std of our time
mean = np.mean(time)
_, ax = plt.subplots(figsize=(15,10))
ax.plot(iteration, time, linestyle="--", marker="o", label=f"Standard Deviation = {np.around(std,1)}\nMean = {mean}")
ax.set_ylabel("Duration (min)", fontsize=14)
ax.set_xlabel("Time", fontsize=14)
ax.set_title("The duration to create resources depend on the Azure API network", fontsize=20, pad=10)
ax.set_xticks(iteration, iteration_label)
ax.legend()
for i, txt in enumerate(time):
    ax.annotate(txt, (iteration[i],time[i]-0.5), ha="center", va='top', color='red', size=15)
```
