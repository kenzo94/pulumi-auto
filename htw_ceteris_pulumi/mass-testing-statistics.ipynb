```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
time = [40, 17, 30]
iteration = ["12:00", "", "20:00"]
```

```python
rng = np.random.default_rng(42)
time = rng.integers(low=5, high=60, size=10)
iteration = np.arange(1,11)
```

```python
std=np.std(time) # std of our time
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(iteration, time, linestyle="--", marker="o")
ax.set_ylabel("time in minute", fontsize=14)
ax.set_xlabel("iteration", fontsize=14)
ax.text(8, 50, f"Standard Deviation = {np.around(std,1)}", bbox=dict(facecolor='red', alpha=0.5), fontsize=15, horizontalalignment='center', verticalalignment='center')
for i, txt in enumerate(time):
    ax.annotate(txt, (iteration[i],time[i]+2), ha="center", va='top', color='red', size=15)
```
